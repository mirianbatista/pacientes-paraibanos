```python
import pandas as pd
import numpy as np 
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
counties = pd.read_csv('procedimentos.csv')
res = counties['NO_MUNICIPIO'].to_numpy()
df = np.unique(res)
d = pd.DataFrame({'name': df})

def addPb(municipio):
    pb = municipio.replace(municipio, municipio + '-Paraíba')
    
    return pb
    
d['name'] = d.apply(lambda row: addPb(row['name']), axis=1)
d
```

```python
geolocator = Nominatim(user_agent="specify_your_app_name_here")

from geopy.extra.rate_limiter import RateLimiter
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=3)
d['location'] = d['name'].apply(geocode)

d['point'] = d['location'].apply(lambda loc: tuple(loc.point) if loc else None)
d
```

```python
def semPb(municipio):
    pb = municipio.replace('-Paraíba','')
    
    return pb
    
d['name'] = d.apply(lambda row: semPb(row['name']), axis=1)
d.to_csv("coordenadas.csv", index=False)
d
```
