```python
import pandas as pd
```

```python
hospitais = pd.read_csv("procedimentos.csv")
hosp = hospitais[~hospitais['NO_MUNICIPIO_RES'].isin(['SEM INFORMAÇÃO'])]
```

```python
coordenadas = pd.read_csv("coordenadas.csv")
coordn = pd.DataFrame(coordenadas.point.str.split(' ',1).tolist(), columns = ['latitude','longitude']).join(coordenadas['name'])

def limpaCoord(coord): 
  coord = coord.split(',')[0]
  coord = coord.replace('(', '')
  coord = coord.replace(',', '')

  return coord


coordn['latitude'] = coordn.apply(lambda row: limpaCoord(row['latitude']), axis=1)
coordn['longitude'] = coordn.apply(lambda row: limpaCoord(row['longitude']), axis=1)
coordn
```

```python
h = pd.merge(hosp, coordn, left_on="NO_MUNICIPIO", right_on="name", how="left")
h
```

```python
distancias = h[['NO_MUNICIPIO', 'NO_MUNICIPIO_RES', 'point']]
distancias
```

```python
flou = h[['NO_MUNICIPIO', 'NO_MUNICIPIO_RES', 'latitude', 'longitude']]
flou.to_csv("flourish.csv", index=False)
flou
```

```python
import numpy as np
np.random.seed(42)

remove_n = 1597020
drop_indices = np.random.choice(flou.index, remove_n, replace=False)
floutest = flou.drop(drop_indices)
floutest
```

```python
floutest.to_csv("floutest.csv", index=False)
```
